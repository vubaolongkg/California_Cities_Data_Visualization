```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
cities['area_total_km2'].sort_values(ascending=False)
```

```python
cities
```

```python
cities.isnull().sum()
```

```python
# Extract Latd = VÄ© Ä‘á»™ and Longd = Kinh Ä‘á»™
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'],cities['area_total_km2']
plt.style.use('default')
plt.figure(figsize=(8,6))
# Plot using Pyplot API
plt.scatter(lon,lat, c=np.log10(population),cmap='viridis', s=area,linewidths=0, alpha=0.5);
plt.axis("equal");
plt.xlabel('longlatitude');
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7)

#Create a legend for cities'sizes
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[], s=area, c='k', alpha=0.4, label=str(area) + 'km$^2$');
plt.legend(labelspacing=1, title='City Area');
plt.title('California Cities: Population and Area Distribution')
```

```python
cities = cities.dropna()
```

```python
cities
```

```python
sns.set_style("whitegrid")
```

```python
# 1. Distribution of city elevations
plt.figure(figsize=(10, 5))
sns.histplot(cities['elevation_m'], bins=30, kde=True, color='blue')
plt.title('Distribution of City Elevations')
plt.xlabel('Elevation (m)')
plt.ylabel('Number of Cities')
plt.show()
```

#### *Elevation Distribution*: If the histogram follows a normal distribution, it indicates that most cities are situated within a certain elevation range.

```python
# 2. Population distribution
plt.figure(figsize=(10, 5))
sns.histplot(cities['population_total'], bins=30, kde=True, color='red')
plt.title('Population Distribution of Cities')
plt.xlabel('Population')
plt.ylabel('Number of Cities')
plt.show()
```

#### *Population Distribution*: The histogram may reveal an imbalance, where many cities have small populations while a few have very large populations.

```python
# 3. Relationship between population and area
plt.figure(figsize=(10, 5))
sns.scatterplot(x=cities['area_total_km2'], y=cities['population_total'], alpha=0.6)
plt.title('Relationship Between Population and City Area')
plt.xlabel('Area (kmÂ²)')
plt.ylabel('Population')
plt.xscale('log')  # Use log scale for better visualization
plt.yscale('log')
plt.show()
```

#### *Population vs. Area*: If there is a linear trend, it suggests that larger cities tend to have higher populations. If there are outliers, some cities may be densely populated despite having a small area.

```python
# 4. Relationship between population and elevation
plt.figure(figsize=(10, 5))
sns.scatterplot(x=cities['elevation_m'], y=cities['population_total'], alpha=0.6, color='green')
plt.title('Relationship Between Population and Elevation')
plt.xlabel('Elevation (m)')
plt.ylabel('Population')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

#### *Population vs. Elevation*: If the population decreases as elevation increases, it suggests that higher-altitude cities tend to have fewer inhabitants.

```python
# 5. Top 10 most populated cities
top_10_cities = cities.nlargest(10, 'population_total')
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cities['population_total'],hue=top_10_cities['population_total'], y=top_10_cities['city'], palette='viridis',legend=False)
plt.title('Top 10 Most Populated Cities')
plt.xlabel('Population')
plt.ylabel('City')
plt.show()
```

#### *Top 10 Most Populated Cities*: The bar chart highlights the cities with the highest populations.

```python
# 6. Proportion of land vs. water area
plt.figure(figsize=(10, 5))
cities[['area_land_sq_mi', 'area_water_sq_mi']].sum().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'navy'])
plt.title('Proportion of Land vs. Water Area')
plt.ylabel('')  # Remove y-label
plt.show()
```

#### *Proportion of Land vs. Water Area*: A pie chart shows how much of the total area consists of land vs. water.


```python
# 7. Latitude vs. Population
plt.figure(figsize=(10, 5))
sns.scatterplot(x=cities['latd'], y=cities['population_total'], alpha=0.6, color='purple')
plt.title('Relationship Between Latitude and Population')
plt.xlabel('Latitude')
plt.ylabel('Population')
plt.show()
```

#### *Latitude vs. Population*: Helps determine if population density is affected by proximity to the equator.
